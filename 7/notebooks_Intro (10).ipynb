```python
import numpy as np
import matplotlib.pyplot as plt


x = np.array([i * 0.05 for i in range(11)])
y = np.array([4.2526, 4.3532, 4.4552, 4.5618, 4.6743, 4.7903, 4.9130, 5.0419, 5.1774, 5.3201, 5.4706])

def divided_differences(x, y):
    n = len(x)
    f = np.zeros((n, n))
    f[:, 0] = y
    for j in range(1, n):
        for i in range(n - j):
            f[i, j] = (f[i + 1, j - 1] - f[i, j - 1]) / (x[i + j] - x[i])
    return f

def newton_interpolation(x, y, x0):
    n = len(x)
    f = divided_differences(x, y)
    ans = f[0, 0]
    term = 1
    for j in range(1, n):
        term *= (x0 - x[j - 1])
        ans += f[0, j] * term
    return ans
x1 = 1.361
x2 = 1.394
y1 = newton_interpolation(x, y, x1)
y2 = newton_interpolation(x, y, x2)
print(f"f({x1}) = {y1}")
print(f"f({x2}) = {y2}")

xx = np.linspace(np.min(x), np.max(x), 100)
yy = np.zeros_like(xx)
for i in range(len(xx)):
    yy[i] = newton_interpolation(x, y, xx[i])

plt.plot(x, y, 'o', label='–î–∞–Ω–Ω—ã–µ —Ç–æ—á–∫–∏')
plt.plot(xx, yy, label='–ë–∞–≥–∞—Ç–æ—à–ª–µ–Ω –ù—å—é—Ç–æ–Ω–∞')
plt.title("–ì—Ä–∞—Ñ–∏–∫ –∏–Ω—Ç–µ—Ä–ø–æ–ª—è—Ü–∏–æ–Ω–Ω–æ–π —Ñ—É–Ω–∫—Ü–∏–∏ –ù—å—é—Ç–æ–Ω–∞")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```
